Implementing K means from scrach 

```python

def initialize_centroids(data, k):
    indices = np.random.choice(range(len(data)), size=k, replace=False)
    centroids = data[indices]
    return centroids

k = 3 
centroids = initialize_centroids(data, k)

def assign_clusters(data, centroids):
    clusters = np.zeros(len(data), dtype=int)
    for i, point in enumerate(data):
        distances = np.linalg.norm(point - centroids, axis=1)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    return clusters

clusters = assign_clusters(data, centroids)

def update_centroids(data, clusters, k):
    centroids = np.zeros((k, data.shape[1]))
    for cluster in range(k):
        cluster_points = data[clusters == cluster]
        centroids[cluster] = np.mean(cluster_points, axis=0)
    return centroids

centroids = update_centroids(data, clusters, k)

def k_means_clustering(data, k, max_iterations=100):
    centroids = initialize_centroids(data, k)
    for _ in range(max_iterations):
        old_centroids = centroids
        clusters = assign_clusters(data, centroids)
        centroids = update_centroids(data, clusters, k)
        if np.all(old_centroids == centroids):
            break
    return clusters, centroids

clusters, centroids = k_means_clustering(data, k)



predicted_labels_knn = np.zeros_like(clusters)
for cluster in range(k):
    cluster_label = np.argmax(np.bincount(target[clusters == cluster]))
    predicted_labels_knn[clusters == cluster] = cluster_label


```

```python
def calculate_accuracy(y_true, y_pred):
    correct_predictions = np.sum(y_true == y_pred)
    total_predictions = len(y_true)
    accuracy = correct_predictions / total_predictions
    return accuracy
```

```python
calculate_accuracy(target, predicted_labels_knn)
```

# K means with Principle component Analysis

```python


def pca(X, k):
    X_std = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    cov_matrix = np.cov(X_std.T)
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    idx = eigenvalues.argsort()[::-1]  
    eigenvectors = eigenvectors[:, idx]
    X_pca = np.dot(X_std, eigenvectors[:, :k])
    print("eigenvalues = ",eigenvalues )
    print("eigenvectors = ", eigenvectors)
    return X_pca


k = 1 

X_pca = pca(data, k)
```
